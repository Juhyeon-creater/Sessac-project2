```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import matplotlib as mpl

font_path = "/content/NanumGothic.ttf"
font_prop = fm.FontProperties(fname=font_path)

# 폰트 등록
fm.fontManager.addfont(font_path)
font_name = font_prop.get_name()
print("폰트 이름:", font_name)

# 전역 적용 (되면 좋고, 안 되면 개별 적용으로 커버됨)
mpl.rcParams['font.family'] = font_name
mpl.rcParams['axes.unicode_minus'] = False


# 1. CSV LOAD
csv_path = "/content/drive/MyDrive/shared_googledrive(Sessac Final Project)/video/머메이드/HJ_Hundred_20251120 편집.csv"
df = pd.read_csv(csv_path)

# YOLO 다리 각도 절댓값
df["Leg_Angle"] = df["Angle"].abs()

# 골반 기준점 보정 후 절댓값(센서 기반 흔들림)
baseline = -178.943
df["RollDiff"] = (df["MPU_Roll_Diff"] - baseline).abs()

# 2. 센서 기반 라벨 생성 (Ground Truth)
#    RollDiff > 5°  → 비정상 = 1, 정상 = 0
SENSOR_THRESHOLD = 5
df["Shake_Flag"] = df["RollDiff"].apply(lambda x: 1 if x > SENSOR_THRESHOLD else 0)

print("\n[라벨 분포]")
print(df["Shake_Flag"].value_counts())

target_col = "Shake_Flag"

# 3. 피처셋 정의

# A. YOLO-only (센서 없이)
feature_cols_yolo = ["Leg_Angle"]

# B. YOLO + 센서
feature_cols_yolo_sensor = [
    "Leg_Angle", "RollDiff",
    "MPU1_Roll", "MPU2_Roll",
    "M1_AcX", "M2_GyY"
]

# 공통으로 쓸 데이터프레임 (결측치 제거)
df_ml = df.dropna(subset=feature_cols_yolo_sensor + [target_col])

X_A = df_ml[feature_cols_yolo].values          # YOLO-only
X_B = df_ml[feature_cols_yolo_sensor].values   # YOLO+Sensor
y = df_ml[target_col].values

# 4. Train/Test Split (두 모델 동일 split)
X_train_A, X_test_A, y_train, y_test = train_test_split(
    X_A, y, test_size=0.2, random_state=42, stratify=y
)

# 같은 random_state, 같은 y → 같은 인덱스로 나뉨
X_train_B, X_test_B, _, _ = train_test_split(
    X_B, y, test_size=0.2, random_state=42, stratify=y
)

# 5. Logistic / RandomForest 평가 함수
def evaluate_baseline(X_train, X_test, y_train, y_test, title="Model"):
    print(f"\n {title} ")

    # Logistic Regression
    lr = LogisticRegression(max_iter=500)
    lr.fit(X_train, y_train)
    pred_lr = lr.predict(X_test)

    print("\n Logistic Regression ")
    print(classification_report(y_test, pred_lr))
    print(confusion_matrix(y_test, pred_lr))

    # Random Forest
    rf = RandomForestClassifier(n_estimators=200, random_state=42)
    rf.fit(X_train, y_train)
    pred_rf = rf.predict(X_test)

    print("\n Random Forest ")
    print(classification_report(y_test, pred_rf))
    print(confusion_matrix(y_test, pred_rf))

# Baseline 비교
evaluate_baseline(X_train_A, X_test_A, y_train, y_test, "YOLO Only (Baseline)")
evaluate_baseline(X_train_B, X_test_B, y_train, y_test, "YOLO + Sensor (Baseline)")

# 6. LSTM용 시계열 데이터 생성
SEQ_LEN = 20

def make_sequence(feature_array, labels, seq_len=20):
    """
    feature_array: (N, D)
    labels: (N,)
    return: X_seq (N-seq_len, seq_len, D), y_seq (N-seq_len,)
    """
    X_seq, y_seq = [], []
    for i in range(len(feature_array) - seq_len):
        X_seq.append(feature_array[i:i+seq_len])
        y_seq.append(labels[i+seq_len])
    return np.array(X_seq), np.array(y_seq)

# 스케일링 후 시계열 생성
scaler_A = MinMaxScaler()
X_A_scaled = scaler_A.fit_transform(X_A)
X_seq_A, y_seq = make_sequence(X_A_scaled, y, SEQ_LEN)

scaler_B = MinMaxScaler()
X_B_scaled = scaler_B.fit_transform(X_B)
X_seq_B, y_seq_B = make_sequence(X_B_scaled, y, SEQ_LEN)

# y_seq와 y_seq_B는 동일해야 함
assert np.array_equal(y_seq, y_seq_B)
y_seq = y_seq_B  # 그냥 하나만 사용

# 7. 시계열 Train/Test Split (두 모델 공통 인덱스)
indices = np.arange(len(y_seq))
idx_train, idx_test = train_test_split(
    indices, test_size=0.2, random_state=42, stratify=y_seq
)

X_train_A_lstm = X_seq_A[idx_train]
X_test_A_lstm  = X_seq_A[idx_test]
X_train_B_lstm = X_seq_B[idx_train]
X_test_B_lstm  = X_seq_B[idx_test]
y_train_lstm   = y_seq[idx_train]
y_test_lstm    = y_seq[idx_test]

# 8. LSTM 모델 정의 & 학습 함수
def build_lstm(input_dim):
    model = Sequential([
        LSTM(64, input_shape=(SEQ_LEN, input_dim)),
        Dropout(0.3),
        Dense(32, activation="relu"),
        Dense(1, activation="sigmoid")
    ])
    model.compile(
        loss="binary_crossentropy",
        optimizer=tf.keras.optimizers.Adam(1e-3),
        metrics=["accuracy"]
    )
    return model

def train_and_eval_lstm(X_train, X_test, y_train, y_test, title="LSTM"):
    model = build_lstm(X_train.shape[2])
    history = model.fit(
        X_train, y_train,
        epochs=20,
        batch_size=16,
        validation_split=0.2,
        verbose=1
    )

    pred_prob = model.predict(X_test)
    pred = (pred_prob > 0.5).astype(int)

    print(f"\n {title} ")
    print(classification_report(y_test, pred))
    print(confusion_matrix(y_test, pred))

    return model, history


# YOLO-only LSTM
model_A, history_A = train_and_eval_lstm(
    X_train_A_lstm, X_test_A_lstm, y_train_lstm, y_test_lstm,
    title="LSTM - YOLO Only"
)

# YOLO+Sensor LSTM
model_B, history_B = train_and_eval_lstm(
    X_train_B_lstm, X_test_B_lstm, y_train_lstm, y_test_lstm,
    title="LSTM - YOLO + Sensor"
)

# 9. Loss / Accuracy 그래프 (두 모델 비교)
plt.figure(figsize=(12, 8))

# Loss
plt.subplot(2, 2, 1)
plt.plot(history_A.history["loss"], label="Train Loss (YOLO)")
plt.plot(history_A.history["val_loss"], label="Val Loss (YOLO)")
plt.title("Loss Curve - YOLO Only")
plt.xlabel("Epoch"); plt.ylabel("Loss")
plt.grid(); plt.legend()

plt.subplot(2, 2, 2)
plt.plot(history_B.history["loss"], label="Train Loss (YOLO+Sensor)")
plt.plot(history_B.history["val_loss"], label="Val Loss (YOLO+Sensor)")
plt.title("Loss Curve - YOLO + Sensor")
plt.xlabel("Epoch"); plt.ylabel("Loss")
plt.grid(); plt.legend()

# Accuracy
plt.subplot(2, 2, 3)
plt.plot(history_A.history["accuracy"], label="Train Accuracy (YOLO)")
plt.plot(history_A.history["val_accuracy"], label="Val Accuracy (YOLO)")
plt.title("Accuracy Curve - YOLO Only")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.grid(); plt.legend()

plt.subplot(2, 2, 4)
plt.plot(history_B.history["accuracy"], label="Train Accuracy (YOLO+Sensor)")
plt.plot(history_B.history["val_accuracy"], label="Val Accuracy (YOLO+Sensor)")
plt.title("Accuracy Curve - YOLO + Sensor")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.grid(); plt.legend()

plt.tight_layout()
plt.show()
```
