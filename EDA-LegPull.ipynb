```python
import pandas as pd

df = pd.read_csv("dataset/LegPull/LegPull -_cam1/keypoints_필라테스_가산_A__Leg Pull_고급_actorP061_20220829_11.23.50_CAM_1.csv")
df
```

```python
df.columns
```

```python
joint_names = {
    'Head': ('Head_x', 'Head_y'),
    'Neck': ('Neck_x', 'Neck_y'),
    'LShoulder': ('LShoulder_x', 'LShoulder_y'),
    'RShoulder': ('RShoulder_x', 'RShoulder_y'),
    'LElbow': ('LElbow_x', 'LElbow_y'),
    'RElbow': ('RElbow_x', 'RElbow_y'),
    'LWrist': ('LWrist_x', 'LWrist_y'),
    'RWrist': ('RWrist_x', 'RWrist_y'),
    'LHip': ('LHip_x', 'LHip_y'),
    'RHip': ('RHip_x', 'RHip_y'),
    'LKnee': ('LKnee_x', 'LKnee_y'),
    'RKnee': ('Rknee_x', 'Rknee_y'),
    'LAnkle': ('LAnkle_x', 'LAnkle_y'),
    'RAnkle': ('RAnkle_x', 'RAnkle_y')
}
connections = [
    ('Head', 'Neck'),
    ('Neck', 'LShoulder'), ('Neck', 'RShoulder'),
    ('LShoulder', 'LElbow'), ('LElbow', 'LWrist'),
    ('RShoulder', 'RElbow'), ('RElbow', 'RWrist'),
    ('LShoulder', 'LHip'), ('RHip', 'RShoulder'), # 필라테스에 맞춰 쇄골-힙 연결
    ('LHip', 'LKnee'), ('LKnee', 'LAnkle'),
    ('RHip', 'RKnee'), ('RKnee', 'RAnkle'),
    ('LHip', 'RHip') # 골반 연결
]
```

```python
import matplotlib.animation as animation
from PIL import Image
import numpy as np


fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(df[[col for col in df.columns if '_x' in col]].min().min() - 50, 
            df[[col for col in df.columns if '_x' in col]].max().max() + 50)
ax.set_ylim(df[[col for col in df.columns if '_y' in col]].min().min() - 50, 
            df[[col for col in df.columns if '_y' in col]].max().max() + 50)
ax.invert_yaxis() # 이미지 좌표계는 Y축이 위에서 아래로 증가
ax.set_title("Pilates Pose Animation")
ax.set_xlabel("X Coordinate")
ax.set_ylabel("Y Coordinate")
ax.grid(True)

# 초기 뼈대 (빈 그래프)
line, = ax.plot([], [], 'o-', lw=2)
time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)


# 애니메이션 초기화 함수
def init():
    line.set_data([], [])
    time_text.set_text('')
    return line, time_text

# 각 프레임을 그리는 함수
def animate(i):
    current_frame_data = df.iloc[i]
    
    x_data = []
    y_data = []
    
    # 관절점 그리기
    for joint_name, (x_col, y_col) in joint_names.items():
        x_data.append(current_frame_data[x_col])
        y_data.append(current_frame_data[y_col])
    
    # 뼈대 연결
    all_x = []
    all_y = []
    for conn in connections:
        j1_name, j2_name = conn
        x1, y1 = current_frame_data[joint_names[j1_name][0]], current_frame_data[joint_names[j1_name][1]]
        x2, y2 = current_frame_data[joint_names[j2_name][0]], current_frame_data[joint_names[j2_name][1]]
        
        all_x.extend([x1, x2, np.nan]) # np.nan으로 선 끊기
        all_y.extend([y1, y2, np.nan]) # np.nan으로 선 끊기

    line.set_data(all_x, all_y)
    time_text.set_text(f'Frame: {i}')
    return line, time_text

# 4. 애니메이션 생성 및 저장
# interval: 각 프레임 사이 간격 (ms) -> 33ms는 약 30fps
# frames: 총 프레임 수
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=len(df), interval=33, blit=True)

# MP4 파일로 저장 (ffmpeg 설치 필요)
# ani.save('pilates_pose_animation.mp4', writer='ffmpeg', fps=30) 

#GIF 파일로 저장 (ImageMagick 또는 Pillow 설치 필요)
ani.save('pilates_pose_animation.gif', writer='pillow', fps=30) 

plt.show()
```
